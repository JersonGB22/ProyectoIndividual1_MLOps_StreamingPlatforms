Por lo tanto, aprovechando esta situación se crea una función que reciba un data frame y una letra (para la transformación 1) y devuelva el data frame con solo las transformaciones pedidas.

### Transformaciones

```python
def transformation(df,lt):

    #Transformación 1: generación del campo "id"
    df.insert(0,"id",lt+df.show_id.astype(str))
    df.drop(columns=["show_id"],inplace=True)

    #Transformación 2: reemplazo de valores nulos del campo rating por el string “G”
    df.rating.fillna("G",inplace=True)

    #Transformación 3: Convertir las fechas al formato "AAAA-mm-dd"
    df["date_added"]=pd.to_datetime(df.date_added.str.strip(),format="%B %d, %Y")

    #Transformación 4: Convertir los campos de texto a minúscula, sin excepción
    df=df.apply(lambda x: x.str.lower() if x.dtype=="O" else x)

    #Transformación 5: Convertir el campo duration a los campos: duration_int y duration_type
    df["duration_int"]=df.duration.apply(lambda x: int(x.split()[0]) if pd.notnull(x) else x)
    df["duration_type"]=df.duration.apply(lambda x: x if pd.isnull(x) else (x.split()[1][:-1] if x[-1]=="s" else x.split()[1]))
    df.drop(columns=["duration"],inplace=True)

    return df
```

Aplicando la función "transformation" en los data frames y concatenándolos en un data frame general

```python
df_platform=pd.concat([transformation(df_amazon,"a"),
                       transformation(df_disney,"d"),
                       transformation(df_hulu,"h"),
                       transformation(df_netflix,"n")]).reset_index(drop=True)
df_platform.head(3)
```

```python
df_platform.info()
```

### Determinación de los puntajes promedio de cada contenido audiovisual

Carga y exploración de datos

```python
df1=pd.read_csv(r"Datasets/ratings/1.csv")
df2=pd.read_csv(r"Datasets/ratings/2.csv")
df3=pd.read_csv(r"Datasets/ratings/3.csv")
df4=pd.read_csv(r"Datasets/ratings/4.csv")
df5=pd.read_csv(r"Datasets/ratings/5.csv")
df6=pd.read_csv(r"Datasets/ratings/6.csv")
df7=pd.read_csv(r"Datasets/ratings/7.csv")
df8=pd.read_csv(r"Datasets/ratings/8.csv")
```

```python
df1.head(3)
```

```python
df1.info()
```

* La conclusión de que todos los otros 7 data frames tienen la misma cantidad de columnas, con el mismo nombre, del mismo tipo y sin ningún valor nulo, se deja al análisis del lector.

Concatemos los data frames y creamos uno que tenga una columna con puntaje promedio (score) por cada movieId.

```python
df_rating=pd.concat([df1,df2,df3,df4,df5,df6,df7,df8])
df_score=df_rating.groupby("movieId")["rating"].mean().round(2).to_frame().reset_index()
df_score.rename(columns={"rating":"score"},inplace=True)
df_score
```

Agregamos la columna "score" al data frame "df_platform", respetando los "id"

```python
# Se elimina la columna "movieId" por redundancia
df_platform=pd.merge(df_platform,df_score,left_on="id",right_on="movieId",how="inner").drop(columns=["movieId"])
df_platform.head(3)
```
